#### Задача 4.8



1. Чему равен ранг полученной корреляционной матрицы?
2. Чему равен определитель полученной корреляционной матрицы? Ответ округлите до седьмого знака после точки-разделителя.
3. Что можно сказать об обусловленности корреляционной матрицы?

````python
def make_vec_st(vec: np.array) -> np.array:
    for it in vec:
        if isinstance(it, type(np.array([]))):
            raise ValueError("I work with vecs only")
    vec_st = (vec - vec.mean())/np.linalg.norm(vec - vec.mean())
    return vec_st
````

````python
x_1 = np.array([5.1, 1.8, 2.1, 10.3, 12.1, 12.6])
x_2 = np.array([10.2, 3.7, 4.1, 20.5, 24.2, 24.1])
x_3 = np.array([2.5, 0.9, 1.1, 5.1, 6.1, 6.3])
# np
x_1_st = make_vec_st(x_1)
x_2_st = make_vec_st(x_2)
x_3_st = make_vec_st(x_3)
x_matrix = np.matrix([x_1_st, x_2_st, x_3_st])
corr_np_matrix = np.corrcoef(x_matrix, rowvar=True)
# 1
display(np.linalg.matrix_rank(corr_np_matrix))
# 2
display(np.linalg.det(corr_np_matrix))

# pd
df = pd.DataFrame({"x_1": x_1, "x_2": x_2, "x_3": x_3})
corr_pd_matrix = df.corr()
# 1
display(corr_pd_matrix.rank())
# 2
display(np.linalg.det(corr_pd_matrix))
````
