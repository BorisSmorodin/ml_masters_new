## Задание 1
Исследуйте систему векторов на линейную зависимость или независимость без использования Python (вручную):

а1 = (1, 8, -1)
а2 =  (-2, 3, 3)
a3 = (4, -11, 9)  

Подробно распишите решение, а в итоге сделайте вывод про то, являются ли векторы зависимыми или независимыми

***Определение линейной зависимости***:

Векторы $a_1, a_2, a_3$ линейно зависимы, если существуют такие коэффициенты $w_1, w_2, w_3$, совместно не равные нулю, что выполняется следующее условие:

$
w_1 a_1 + w_2 a_2 + w_3 a_3 = 0.
$

Если таких коэффициентов нет, то векторы линейно независимы.

---
***1. Составим уравнения:***

Подставим векторы $a_1, a_2, a_3$:

$
w_1 (1, 8, -1) + w_2 (-2, 3, 3) + w_3 (4, -11, 9) = (0, 0, 0).
$

Получим систему линейных уравнений:

$
\begin{aligned}
& \quad w_1 - 2w_2 + 4w_3 = 0, \\
& \quad 8w_1 + 3w_2 - 11w_3 = 0, \\
& \quad -w_1 + 3w_2 + 9w_3 = 0.
\end{aligned}
$


***2. Запишем систему в матричном виде:***

$$
\begin{bmatrix}
1 & -2 & 4 \\
8 & 3 & -11 \\
-1 & 3 & 9
\end{bmatrix}
\cdot 
\begin{bmatrix}
w_1 \\
w_2 \\
w_3
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0
\end{bmatrix}
$$

Будем искать нетривиальные решения данной системы. Чтобы определить, существуют ли такие решения, необходимо найти ранг составленной матрицы матрицы

***3. Находим ранг матрицы***
Найдем определитель:

$
\text{det}
\begin{bmatrix}
1 & -2 & 4 \\
8 & 3 & -11 \\
-1 & 3 & 9
\end{bmatrix}
$


$$
\text{det} =
1 \cdot
\begin{vmatrix}
3 & -11 \\
3 & 9
\end{vmatrix}
-
(-2) \cdot
\begin{vmatrix}
8 & -11 \\
-1 & 9
\end{vmatrix}
+
4 \cdot
\begin{vmatrix}
8 & 3 \\
-1 & 3
\end{vmatrix}
$$

$
\text{det} = 1 \cdot 60 - (-2) \cdot 61 + 4 \cdot 27 = 60 + 122 + 108 = 290
$


Так как определитель матрицы $ \neq 0 $, ранг матрицы равен 3 (максимальный). Следовательно, система уравнений имеет только тривиальное решение $w_1 = w_2 = w_3 = 0$


***Вывод:***

Векторы $a_1 = (1, 8, -1)$, $a_2 = (-2, 3, 3)$, $a_3 = (4, -11, 9)$ линейно независимы

## Задание 10
Разделите данные на предикторы и целевую переменную. Обучите модель с помощью классической линейной регрессии. Линейную регрессию необходимо реализовать без готовых библиотек.

Оцените качество модели с помощью хотя бы двух метрик качества.

Распишите решение с помощью кода на языке Python и напишите текстом интерпретацию полученного вывода.

```python
X = data.drop(columns=["cnt"])  # все переменные, кроме целевой переменной
y = data["cnt"]  # целевая переменная
```

```python
# Формируем bias
X = np.c_[np.ones(X.shape[0]), X]

X = np.array(X)
y = np.array(y)

# находим коэффициенты линейной регрессии
w_hat = np.linalg.inv(X.T @ X) @ (X.T @ y)

display(w_hat)
```

```python
# получаем прогнозы
y_pred = X @ w_hat
display(y_pred)
display(y)
```

```python
# Средняя абсолютная ошибка
mae = np.mean((y - y_pred))

# Средняя абсолютная ошибка в процентах
mape = np.mean((abs(y - y_pred)/y)*100)

print(f"MAE: {mae}")
print(f"MAPE: {mape}")
```

***Вывод:***

Как видно, полученные результаты не являются точными, поскольку средняя относительная ошибка составляет примерно 45%. Скорее всего, нормализация признаков смогла бы улучшить качество результатов работы модели.
