## Задание 1 (4 балла)
Обратный порядок слов в блоках текста

Дан текстовый файл, каждое предложение которого занимает одну строку. Напишите программу, которая разделяет текст на блоки — каждый блок состоит из нескольких предложений. Затем переворачивает порядок слов только внутри каждого предложения, не меняя порядок самих предложений в блоке.

<b>Файл содержит:</b>

Привет как дела<br>
На улице идет дождь<br>
Я люблю программирование<br>

Критерии оценки:

- Текст разделен на блоки корректно — 2 балла.
- Корректно перевернуты блоки — 2 балла.

```python
# Поскольку в задании не указано, сколько предложений должно быть в одном блоке, возьмём за основу 2.
sliding_window: int = 2
# Считываем файл
with open("data/task_1.txt", "r", encoding="utf-8") as f:
    # Считываем строки и удаляем символ переноса строки
    lines = list(map(str.strip, f.readlines()))
# Производим обработку. Заводим счётчик, который будет служить ключом в результирующем словаре. Заводим пустой словарь.
cnt = 1
blocks = {}
# Итерируемся по строкам текста с шагом, равным sliding_window
for i in range(0, len(lines), sliding_window):
    # На каждой итерации берём срез от исходного индекса до его суммы со значением sliding_window
    # Разбиваем каждое предложение по словам и меняем порядок слов, после чего записываем результат в словарь и прибавляем 1 к счётчику
    blocks.update({cnt: [" ".join(sentence.split()[::-1]) for sentence in lines[i: i+sliding_window]]})
    cnt += 1
# Выводим результат
print(blocks)
        
```

## Задание 2 (2 балла)
Напишите функцию, которая принимает строку и сжимает её определённым образом. Строки содержат произвольные символы, включая пробелы и спецсимволы, и требуют точного учета длины при кодировании. Сжатие строки происходит сериями одинаковых символов в формате <i><b>символ + количество</b></i>, но только если длина сжатой строки не превышает исходную.

Примеры:

- вход: aaabbc → вывод: a3b2c,
- вход: abcd → вывод: abcd (так как сжатая равна).

```python
def encode(s: str) -> str:
    # Построим решение на словаре
    dct = {}
    # Считаем количество вхождений символа в строку
    for c in s:
        dct[c] = dct.get(c, 0) + 1
    # Кодируем строку согласно условию
    return "".join(list(map(lambda x: f"{x[0]}{x[1]}" if x[1] > 1 else f"{x[0]}", dct.items())))
# Проверяем на тестовых данных
print(encode("aaabbc"))
print(encode("abcd"))
```

## Задание 3 (4 балла)
Хаотичные скобки
Реализуйте функцию, которая проверяет, правильно ли расставлены скобки в строке (включая круглые, квадратные и фигурные скобки).

Критерии оценки:

- Функция выполняет условие — 4 балла.

```python
def chaotic_brackets(s: str) -> bool:
    """
    Функция, которая определяет правильность расстановки скобок, используя FIFO (стек)
    """
    # Заводим пустой стэк
    stack = []
    # Заводим словарь для хранения пар скобок различных видов
    pairs = {"{": "}", "[": "]", "(": ")"}
    # Итерируемся по исходной строке
    for c in s:
        # Если встретилась открывающая скобка, добавляем её в стек
        if c in pairs:
            stack.append(c)
        # Если встретилась закрывающая скобка и она соответствует своей паре, то удаляем из стэка открывающую скобку
        elif c in pairs.values() and c == pairs[stack[-1]]:
            stack.pop()
        # В ином случае можно утверждать, что скобки расставлены неправильно
        else:
            break
    # Главное условие - пустой стек
    if not stack:
        return True
    else:
        return False

# Тестируем
assert chaotic_brackets("{(([]))}") is True
assert chaotic_brackets("{}([]())") is True
assert chaotic_brackets("") is True
assert chaotic_brackets("2*(1+2)*[{1, 2, 3} ** (alpha/(3+1/2))]") is True
assert chaotic_brackets("{[}]") is False
```

Генератор случайных паролей
Напишите функцию, которая генерирует пароль заданной длины. В реализации надо учитывать, что:

Пароль должен содержать буквы, цифры и специальные символы.
Длина пароля задается пользователем.

Критерии оценки:

- Функция выполняет все заявленные условия — 4 балла.

```python
# Для решения задачи воспользуемся коллекциями символов из модуля string
from string import ascii_letters, digits, punctuation
from random import choice


def pass_gen(password_length: int) -> str:
    """
    Функция, генерирующая пароль заданной пользователем длины
    """
    # Защита от дурака
    if password_length <= 0:
        raise ValueError("Password length must be more than 1")
    # Создаём пароль
    all_characters_list = ascii_letters + digits + punctuation
    return "".join([choice(all_characters_list) for i in range(password_length)])

# Тестируем
print(pass_gen(20))
    
```

## Задание 5 (6 баллов)
Эмуляция работы электронной очереди
Напишите класс для симуляции работы электронной очереди, например, в банке.

Критерии оценки:

- Система должна поддерживать добавление клиентов с указанием их приоритета (например, VIP, обычный) — 1 балл.
- Выбор клиента для следующей обработки должен учитывать приоритет — 1 балл.
- Реализуйте отчет для администрации с информацией, сколько времени заняло обслуживание клиентов — 4 балла.

```python
from enum import Enum
from random import randrange


class Priority(Enum):
    """
    Класс, отражающий приоритет клиентов
    """
    VIP = 1
    STANDARD = 0   
    

class Client:
    """
    Класс, отражающий сущность клиента.
    Пример создания:
    Client(priority=Priority.VIP)
    """
    ID: int = -1
    
    def __new__(cls, *args, **kwargs):
        """
        Генерируем только уникальные ID клиентов на уровне класса
        """
        cls.ID += 1
        return super().__new__(cls)
        
        
    def __init__(self, priority: Priority):
        self.__id: int = self.ID 
        self.__priority: Priority = priority
    
        
    @property
    def get_id(self) -> int:
        """
        Геттер для ID клиента
        """
        return self.__id
    
    
    @property
    def priority(self) -> Priority:
        """
        Геттер для Приоритета клиента
        """
        return self.__priority
    
    
    @priority.setter
    def priority(self, priority: Priority):
        """
        Сеттер для Приоритета клиента
        """
        self.priority = priority
        
    def __str__(self):
        return(f"Client ID: {self.__id}, priority:")
        
    
    @staticmethod
    def valid_priority_type(priority):
        """
        Метод проверки валидности типа приоритета клиента
        """
        if not isinstance(priority, Priority):
            raise ValueError("priority value must be Priority ENUM object")
        
        
    def __setattr__(self, name, value):
        """
        Запрет на введение новых аттрибутов для клиента, внедрение проверки валидности типа приоритета
        """
        match name:
            case "_Client__priority":
                self.valid_priority_type(value)
                super().__setattr__(name, value)
            case "_Client__id":
                super().__setattr__(name, value)
            case _:
                raise ValueError(f"You can not create new attributes: {name} {value}")


class QueueNode:
    """
    Служебный класс, отражающий элемент очереди.
    Необходим для назначения ссылок на последующие элементы очереди. Ссылки храняться в атрибуте next
    """
    def __init__(self, client: Client):
        self.valid_client_type(client)
        self.__client = client
        # предполагаемое время обработки клиента
        self.__predict_serve_time = randrange(1, 11)
        self.next = None
        
    @property
    def client(self) -> Client:
        return self.__client
    
    @property
    def serve_time(self) -> int:
        return self.__predict_serve_time
    
    @staticmethod
    def valid_client_type(client: Client):
        if not isinstance(client, Client):
            raise ValueError("client attribute must be Client type")
        
    def __str__(self):
        return f"Queue Node: Client ID {self.client.get_id}, Client Priority {self.client.priority}, Predict serve time {self.serve_time} s"

class Queue:
    """
    Базовый служебный класс очереди, определяющий все основные методы для взаимодействия с очередью
    Экземпляр данного класса обладает 2-мя служебными атрибутами: _head и _tail, указывающими на начало и конец очереди соответственно.
    Также экземпляр данного класса обладает атрибутом _info, в котором хранится отчёт об обработанных клиентах
    """
    def __init__(self):
        self._head: QueueNode | None = None
        self._tail: QueueNode | None = None
        self._info: list[dict[str, str | int]] = []
        
        
    
    def add(self, node: QueueNode):
        """
        Метод добавления элемента в конец очереди
        """
        if not self._head:
            self._head, self._tail = node, node 
        else:
            self._tail.next = node
            self._tail = node
            
    def serve(self) -> QueueNode | None:
        """
        Метод удлаения элемента из начала очереди
        """
        if self._head:
            buf_head = self._head
            self._head = self._head.next
            
            if not self._head:
                self._tail = self._head
            # добавляем в отчёт только что обработанного клиента
            self._info.append(
                {
                    "id": buf_head.client.get_id,
                    "priority": buf_head.client.priority.name,
                    "serve_time": buf_head.serve_time
                }
            )
            return buf_head
        else:
            return None
    
    # Набор геттеров и сеттеров для различных атрибутов
    @property
    def head(self) -> QueueNode | None:
        return self._head
    
    @property
    def tail(self) -> QueueNode | None:
        return self._tail
    
    @property
    def info(self) -> list[dict[str, str | int]]:
        return self._info
            
    def peek(self) -> QueueNode | None:
        return self._tail
    
    def refresh_info(self):
        self.info = []


class VipQueue(Queue):
    """
    Дочерний класс очереди для взаимодействия только с VIP клиентами
    """
    def add(self, node: QueueNode):
        if node.client.priority != Priority.VIP:
            raise ValueError("Only VIP clients can be in VIP Queue")
        super().add(node)
        
class StandardQueue(Queue):
    """
    Дочерний класс очереди для взаимодействия только с STANDARD клиентами
    """
    def add(self, node: QueueNode):
        if node.client.priority != Priority.STANDARD:
            raise ValueError("Only STANDARD clients can be in STANDARD Queue")
        super().add(node)
        
        
class QueueInterface:
    """
    Верхнеуровневый класс, который позволяет менеджить очередь.
    Поскольку наличие прироритетов клиентов равносильно наличию нескольких очередей, реализация очереди с приоритетом в данной задаче именно такова
    Инициализировать экземпляр данного класса также возможно на основе коллекции (списка) клиентов
    """
    def __init__(self, client_list: list[Client] = None):
        # При инициилазции экземпляра создаются также экземпляры служебных классов очередей для различных типов клиентов
        self.__vip_queue = VipQueue()
        self.__standard_queue = StandardQueue()
        if client_list:
            for client in client_list:
                node = QueueNode(client)
                self.__add_node(node)

                
    def __add_node(self, node: QueueNode):
        """
        Служебный метод, который выбирает, в какую очередь добавить элемент очереди в зависимости от приоритета клиента
        """
        if node.client.priority == Priority.VIP:
            self.__vip_queue.add(node)
        else:
            self.__standard_queue.add(node)
                
    def add_client(self, client: Client):
        """
        Метод добавляет в соответствующую очередь 1-го клиента
        """
        node = QueueNode(client)
        self.__add_node(node)
        
        
    def serve(self) -> QueueNode | None:
        """
        Метод удаляет из очереди клиента в порядке приоритета
        """
        result = self.__vip_queue.serve()
        if not result:
            result = self.__standard_queue.serve()
        return result
    
    def info(self) -> dict[str, list[dict[str, str | int]]]:
        """
        Геттер, возвращающий отчёт об обработанных клиентах в формате json 
        """
        return {"STANDARD": self.__standard_queue.info, "VIP": self.__vip_queue.info}
    
    def refresh_info(self):
        self.__vip_queue.refresh_info()
        self.__standard_queue.refresh_info()
```

```python
# end-to-end тест
import pandas as pd

client_1 = Client(priority=Priority.STANDARD)
client_2 = Client(priority=Priority.VIP)
client_3 = Client(priority=Priority.STANDARD)
client_4 = Client(priority=Priority.VIP)
client_5 = Client(priority=Priority.STANDARD)
client_6 = Client(priority=Priority.STANDARD)

q = QueueInterface([client_1, client_2, client_3])
q.add_client(client_4)
q.add_client(client_5)
q.add_client(client_6)

served = q.serve()
assert served.client.get_id == client_2.get_id
served = q.serve()
assert served.client.get_id == client_4.get_id
served = q.serve()
assert served.client.get_id == client_1.get_id
served = q.serve()
assert served.client.get_id == client_3.get_id
served = q.serve()
assert served.client.get_id == client_5.get_id
client_7 = Client(priority=Priority.VIP)
q.add_client(client_7)
served = q.serve()
assert served.client.get_id == client_7.get_id
served = q.serve()
assert served.client.get_id == client_6.get_id
served = q.serve()
assert served is None

# информационная сводка для аналитиков
info = q.info()
standard_df = pd.DataFrame(info.get("STANDARD"))
display(standard_df)
vip_df = pd.DataFrame(info.get("VIP"))
display(vip_df)



```

## Задание 6 (2 балла)
Проверка на «почти палиндром»
Напишите программу, которая проверяет, является ли строка палиндромом или «почти палиндромом». «Почти палиндром» означает, что можно удалить одну букву, чтобы строка стала палиндромом.

Критерии оценки:

- Функция выполняет все заявленные условия — 2 балла.

```python
def almost_palindrome(s: str) -> bool:
    """
    Функция, которая выполняет проверку строки на то, является ли она палиндромом или почти палиндромом
    """
    # будем решать задачу за O(n^2) как действительно крутые ребята!
    fl = True
    # Проводим проверку на полный палиндром
    for j in range(len(s)//2):
        if s[j] != s[-j-1]:
            break
    else:
        return True
    # Проверка на почти палиндром
    for i in range(len(s)):
        # Выделяем из исходной строки подстроки с исключением одного символа
        potential_palindrome_string = s[:i] + s[i+1:]
        # Проводим проверку на палиндром
        for j in range(len(potential_palindrome_string)//2):
            if potential_palindrome_string[j] != potential_palindrome_string[-j-1]:
                fl = False
                break
        # Если хотя бы один палиндром был найден среди подстрок, возврашаем True
        else:
            fl = True
            break
    return fl
assert almost_palindrome("abccba") is True # Полный палиндром
assert almost_palindrome("acbba") is True # Почти палиндром (при удалении 'c')
assert almost_palindrome("acaba") is False # Не палидром и не почти палиндром
```

## Задание 7 (9 баллов)
Задача Шредингера
Разработайте программу, которая «стирает» фрагменты текста в файле. Например:

Пользователь указывает файл и процент текста, который нужно удалить (например, 30%).
Программа случайно выбирает слова или части абзацев и заменяет их на пробел или ..., сохраняя общий объем документа.

Вход:
- Сегодня солнечный день, и я собираюсь гулять в парке с моими друзьями.
- А я хочу сидеть дома и усердно решать важный экзамен!

Вывод (удалено ~30%):
- Сегодня ... день, и я собираюсь ... в парке с моими друзьями.
- А ... хочу ... дома ... усердно ... важный экзамен!

Критерии оценки:

- Чтение файла и ввод параметров пользователя — 3 балла.
- Реализация алгоритма случайного удаления текста — 5 баллов.
- Запись результата и вывод пользователю — 1 балл.


```python
from random import randrange

def encode_text(filepath: str, encoding_rate: int) -> str:
    """
    Функция, которая производит шифрование заданного процента слов в тексте
    """
    # Считываем файл
    with open(filepath, "r", encoding="utf-8") as f:
        lines = list(map(str.strip, f.readlines()))
    # Выделяем слова в строках
    words = [it.split() for it in lines]
    # Подсчитываем общее количество слов в тексте
    all_words = sum(len(it) for it in words)
    encoding_rate_percent = encoding_rate / 100
    encoded_words_counter = 0
    cur_rate = encoded_words_counter / all_words
    # Производим шифрование слов согласно критерию
    while cur_rate < encoding_rate_percent:
        line_idx = randrange(0, len(words))
        encode_idx = randrange(0, len(words[line_idx]))
        words[line_idx][encode_idx] = "..."
        encoded_words_counter += 1
        cur_rate = encoded_words_counter / all_words
    lines = [" ".join(it) for it in words]
    # Возвращаем результат
    return lines
           
for it in encode_text("data/task_7.txt", 30):
    print(it)
```
